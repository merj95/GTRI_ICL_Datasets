```python
#Total # of unique businesses
len(scores['business_id'].unique())
```

```python
#Types of inspections
print(scores['inspection_type'].unique())
#There are 15 inspection types
```

```python
filtered_scores = scores[scores['inspection_score'].notnull()]
filtered_scores['inspection_type'].unique()
#only routine - unscheduled inspection types are given an inspection score
```

```python
#Look at a single business to see what a series of inspections may look like
print(scores[scores['business_id'] == 86718])
#This example happens to be of a business (94133) that had no violations and scored 100
#This business (86718) had many violations all during the same inspection
#This tells me that an inspection score will be listed multiple times for one inspection of a business if it has multiple 
#violations.  This will need to be accounted for when looking at score summary data.
```

```python
#How much of the data are null values? (Percentage)
(1 - scores.count()/len(scores))*100
```

```python
#Remove duplicates
scores.drop_duplicates('inspection_id', inplace = True)
#Now scores has is set in such a way that even if a business had multiple violations on the same inspection, that inspection is
#only listed once in the dataset
```

```python
#Total number of the different inspections by type
scores['inspection_type'].value_counts()
```

```python
#Summary of all scores
unsched_scores = scores[scores['inspection_type'] == 'Routine - Unscheduled']
boxplot = unsched_scores.boxplot('inspection_score', figsize=(10,10))
```

```python
#Histogram of scores
hist = unsched_scores['inspection_score'].plot.hist(bins=11, figsize=(8,8))
```

```python
#Summary table of scores by zip code
unsched_scores.boxplot(column='inspection_score',by='business_postal_code', figsize=(15,8))
```
